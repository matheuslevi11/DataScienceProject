## Análise de bases ao longo do tempo

importando todas as bases de dados e realizando análises temporais

```python
from script_tratamento_aposentados import tratamento_aposentados

time_series = []
filenames = ['01_2017', '07_2017', '01_2018', '07_2018','01_2019', '06_2019', '05_2020', '07_2021', '11_2021']

for file in filenames:
    df = tratamento_aposentados(file)
    time_series.append(df)
```

```python
df
df.groupby(df.desc_cargo)['valor_aposentadoria'].describe().sort_values('count', axis=0, ascending=False).head(20)
```

Um mesmo cargo normalmente tem variações de salários alta.

```python
means = []
index = [x for x in range(len(time_series))]
for db in time_series:
    means.append(db.shape[0])

sns.lineplot(x=index, y=means)
plt.title("Quantidade de aposentadorias ao longo dos meses")
plt.xlabel('Tempo')
plt.ylabel('Aposentados')
plt.xticks(ticks=[0,2,4,6,8], labels=[2017, 2018, 2019, 2020, 2021])
plt.show()
```

Houve uma tendência de aposentadoria entre 2019 e 2020, seguida de uma tendência de queda.
